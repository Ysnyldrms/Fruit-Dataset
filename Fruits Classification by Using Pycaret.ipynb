```python
from pycaret.classification import * 
```

```python
clf = setup(data=df, target = "Class", session_id=123, normalize= True)
```

```python
compare_models(sort="AUC")
```

```python
lr = create_model("lr", fold=10)
```

```python
gbc = create_model("gbc", fold=10)
```

```python
rf = create_model("rf", fold=10)
```

```python
blend_specific_soft = blend_models(estimator_list = [lr,gbc,rf], method = "soft")
```

```python
plot_model(blend_specific_soft)
```
